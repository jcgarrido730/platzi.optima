```python
def obtener_notas(soup):
    '''
    funcion que recibe un objeto de BeautifulSoup de una pagina de una sección
    y devuelve una lista de URLs a las notas de esa sección
    '''
    lista_notas = []

    #obtengo el articulo promocionado
    featured_article = soup.find('div', attrs={'class':'article-item__content'})
    if featured_article:
        lista_notas.append(featured_article.a.get('href'))

    #obtengo el listado de articulos 
    article_list = soup.find_all('h4', attrs={'class':'h2 is-display-inline title-list'})

    for article in article_list:
        if article.a:
            lista_notas.append(article.a.get('href'))
    return lista_notas
```

```python
lista_notas = obtener_notas(s_seccion)
lista_notas
```

```python
#url_nota = lista_notas[0]
url_nota = 'https://www.pagina12.com.ar/542291-procrear-maggiotti-pisano-y-bucca-entregaron-viviendas-del-d'
url_nota
```

```python
try:
    nota = requests.get(url_nota)
    if nota.status_code==200:
        s_nota = BeautifulSoup(nota.text, 'lxml')
        #extraer el titulo
        titulo = s_nota.find('h2', attrs={'class':'h4'})
        print(titulo.text)

        #extraer la fecha
        #<div class="date modification-date"><span><time datetime="2023-04-20T16:46:52-03:00">20 de abril de 2023 - 16:46</time></span></div>
        #fecha = s_nota.find('div', attrs={'class':'date modification-date'})
        fecha = s_nota.find('span').get('time')
        print(fecha)
except Exception as e:
    print('Error: ')
    print(e)
    print('\n')
```
